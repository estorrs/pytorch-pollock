```r
for (dtype_list in fmap) {
    for (d1 in dtype_list) {
        for (disease_list in d1) {
            train_fp = disease_list[[1]]
            val_fp = disease_list[[2]]

            dtype = strsplit(strsplit(train_fp, "_train")[[1]][[1]], "/")[[1]][[6]]
            disease = strsplit(strsplit(train_fp, "_train")[[1]][[1]], "/")[[1]][[7]]

            print(dtype)
            print(disease)

            train <- readRDS(train_fp)
            val <- readRDS(val_fp)

            seuratfile = extractSeurat(train, exp_slot_name = "counts")
            train_sampTab = seuratfile$sampTab
            train_sampTab$cell_ids = rownames(train_sampTab)
            train_expDat = seuratfile$expDat

            seuratfile = extractSeurat(val, exp_slot_name = "counts")
            val_sampTab = seuratfile$sampTab
            val_expDat = seuratfile$expDat


            class_info<-scn_train(
                stTrain = train_sampTab, expTrain = train_expDat, nTopGenes = 10, nRand = 70,
                nTrees = 1000, nTopGenePairs = 25, dLevel = "cell_type", colName_samp = "cell_ids")
            classRes_val_all = scn_predict(cnProc=class_info[['cnProc']], expDat=val_expDat, nrand = 50)

            txt_fp = paste("/data/pollock/benchmarking/results/singlecellnet/interdataset_with_folds/", dtype, "_", disease, ".txt", sep = "")
            write.table(classRes_val_all, file = txt_fp, sep="\t")
        }
    }
}
```
