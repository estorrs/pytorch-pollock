```r
library(Seurat)
library(SeuratDisk)
```

```r
prefix = '/data/spatial_transcriptomics//breast_dcis_alla_objects/HT339B1_S1H3_U2'
```

```r
obj = readRDS(paste(prefix, '.rds', sep = ''))
obj
```

```r
obj@meta.data
```

```r
obj@assays$Spatial
```

```r
for (col in colnames(obj@meta.data)) {
    if (!is.null(as.numeric(as.character(obj[[col]][1])))) {
        obj@meta.data[[col]] = as.character(obj@meta.data[[col]])
    }
}
```

```r
slot(obj$SCT@SCTModel.list[[1]], 'median_umi') = median(obj$SCT@SCTModel.list[[1]]@cell.attributes$umi)
```

```r
SaveH5Seurat(obj, filename = paste(prefix, '.h5seurat', sep = ''), overwrite = T)
```
